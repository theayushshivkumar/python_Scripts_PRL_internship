```python
# CONTOUR MAP
from astrodendro.analysis import pp_catalog
import numpy as np
from astrodendro import Dendrogram
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.io.fits import getdata
from astropy.wcs import WCS
from astrodendro.analysis import PPStatistic
from astropy import units as u
from astropy.table import Table
from astropy.visualization import simple_norm
from matplotlib.patches import Ellipse

data = fits.getdata(r"C:\Users\Ayush Shivkumar\OneDrive\Desktop\PRL\FITS files\ALMA\G53.11_MM1.ms.cont.SPW0123.clean.image.pbcor.fits")
#image = np.nan_to_num(image, nan=0.0)
d = Dendrogram.compute(data, min_value = 0.004, min_delta = 0.0015, min_npix=7, wcs = WCS(header))
# print(WCS(header))

%matplotlib qt

stat1 = []
for i in d.leaves:
    s1 = PPStatistic(i)
    stat1.append(s1)
print(stat1[9].major_sigma)
###

major_sigma = []
for i in range(len(stat1)):
    MS = stat1[i].major_sigma * 0.06 # Convert pixel to arcsec
    major_sigma.append(MS)
print(len(major_sigma))

###

minor_sigma = []
for i in range(len(stat1)):
    ms = stat1[i].minor_sigma * 0.06 
    minor_sigma.append(ms)
print(len(minor_sigma))

###

position_angle = []
for i in range(len(stat1)):
    pa = stat1[i].position_angle * 3600 # Deg to Arcsec
    position_angle.append(pa)
print(len(position_angle))

###

radius = []
for i in range(len(stat1)):
    r = stat1[i].radius*0.06
    radius.append(r)
print(len(radius))

###

area_exact = []
for i in range(len(stat1)):
    area = stat1[i].area_exact
    area_exact.append(area)
print(len(area_exact))

###

x_cen = []
for i in range(len(stat1)):
    x = stat1[i].x_cen
    x_cen.append(x)
print(len(x_cen))

x_cen_val = []
for i in range(len(stat1)):
    x = stat1[i].x_cen.value
    x_cen_val.append(x)
print(len(x_cen_val))

##

y_cen = []
for i in range(len(stat1)):
    y = stat1[i].y_cen
    y_cen.append(y)
print(len(y_cen))

y_cen_val = []
for i in range(len(stat1)):
    y = stat1[i].y_cen.value
    y_cen_val.append(y)
print(len(y_cen_val))
##


#print(stat1.major_sigma)
#print(stat1.minor_sigma)
#print(stat1.position_angle)

metadata = {}
metadata['data_unit'] = u.Jy / u.beam
metadata['spatial_scale'] =  0.06 * u.arcsec
metadata['beam_major'] =  1.021960791614E-04 * 3600 * u.arcsec # FWHM
metadata['beam_minor'] =  8.545985652341E-05 * 3600 * u.arcsec # FWHM

stat = []
for i in d.leaves:
    s = PPStatistic(i, metadata=metadata)
    stat.append(s)
    
cat = pp_catalog(d, metadata)
cat.pprint(show_unit=True, max_lines=10)

flux = []
for i in range(len(stat)):
    f = stat[i].flux
    flux.append(f)

print(flux)
print('\n')
major_sig = []
for i in range(len(stat)):
    msig = stat[i].major_sigma
    major_sig.append(msig)
    
print(major_sig)
print('\n')

print(major_sigma)
#all_indices = [structure.idx.value for structure in d.all_structures]
idx = []
for structure in d.leaves:
    id = structure.idx
    idx.append(id)





wcs_coord = []

#print("\n")
#print(x_cen)

pixel_coords = np.array([[x_cen_val, y_cen_val]])

coord = WCS(header).pixel_to_world(pixel_coords[:,0], pixel_coords[:,1])


ra = coord.ra.deg
dec = coord.dec.deg

ra=ra.reshape(33)
dec=dec.reshape(33)

table = Table([idx, flux, ra, dec, major_sigma, minor_sigma, position_angle, radius], names=('idx', 'flux','ra', 'dec', 'major sigma', 'minor sigma', 'position angle', 'radius'))

# table.write(r'C:\Users\Ayush Shivkumar\OneDrive\Desktop\PRL\TESastrodendro_catalogue.csv', format='csv', overwrite=True)

p = d.plotter()

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection = WCS(header))
#norm = simple_norm(image, 'linear', percent=99.5)

im = ax.imshow(data, origin='lower', interpolation='nearest', vmin=0, vmax=0.03, cmap='inferno')
cbar = plt.colorbar(im, ax=ax, orientation='vertical', pad=0.05)
cbar.set_label("Flux Density (Jy/beam)", fontsize=12)

indices_chosen = [7, 9, 11, 13, 15, 16, 19, 20]
for row in table:
    idx = row['idx']
    if idx in indices_chosen:
    
        ra = row['ra']
        dec = row['dec']
    
    
        # Plot the index directly using WCS coordinates
        ax.text(ra, dec, str(idx), transform=ax.get_transform('world'), color='black', fontsize=12, ha='left', va='bottom', bbox=dict(facecolor='white', edgecolor='none', alpha=0.7, pad=1))
    
for leaf in d.leaves:
    
    p.plot_contour(ax, structure = leaf, colors = 'cyan')
    stat = PPStatistic(leaf)
    ellipse = stat.to_mpl_ellipse(edgecolor='black', facecolor='none')

    ax.add_patch(ellipse)
    
ax.set_xlim(433, 540)  # X-pixels
ax.set_ylim(408, 586)  # Y-pixels
ax.set_xlabel("RA (J2000)")
ax.set_ylabel("DEC (J2000)")
# Convert the catalog to an Astropy Table
table = Table(cat)
# Save the table as a CSV file
# table.write(r"C:\Users\Ayush Shivkumar\OneDrive\Desktop\PRL\Astrodendro Data\DIRECTpp_catalogue.csv", format='csv', overwrite=True)
plt.savefig(r"C:\Users\Ayush Shivkumar\OneDrive\Desktop\PRL\FITS files\ALMA\astrodendro_image.png", dpi=300, bbox_inches='tight')
```
