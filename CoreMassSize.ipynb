```python
#Core Mass

import numpy as np
import pandas as pd
def BB(freq, Td):
    return (2 * h * freq**3) / ((c**2) * (np.exp(h * freq / (k * Td)) - 1)) # W/ m^2

import numpy as np

# def Mcore(Fv, d, Kv, wavelength, Td):
#     term1 = 0.12
#     term2 = (np.exp(1.439 / (wavelength * (Td / 10))) - 1)
#     term3 = (Kv / 0.01)**-1
#     term4 = Fv
#     term5 = (d / 100)**2
#     term6 = wavelength**3
#     return term1 * term2 * term3 * term4 * term5 * term6

def Mcore(d, Fv, Kv, freq, Td):
    return ((100 * (d**2) * (Fv))/(Kv * BB(freq, Td))) / solar_mass

data = pd.read_csv(r"C:\Users\Ayush Shivkumar\OneDrive\Desktop\PRL\Astrodendro Data\core_catalogue.csv")
print(data)
# Fluxes of Sources of Interest from Astrodendro Catalogue
Fv = (data['flux (Jy)'])* 10**(-26) # multiply * 10**(-26) for conversion from Jy to W / m^2
# Fv = 2 * 10**(-26) #From paper just to check
#print(Fv)
wavelength = 0.88 # in mm
Td = 15 #Kelvin
beta = 2
# Kv = 0.1 * ((wavelength / 0.30)**(-beta)) * (0.1) # 0.1 for conversion from cm^2 / g to m2 / Kg
Kv = 1.16 * 0.1
print(Kv)
# Kv = 0.18* (0.1) # cm^2/g to m^2/kg
print(Kv)
d = 1700 * 3.086 * 10**16 # in pc ; multiply * 3.086 * 10**16 for pc to m
freq = 340.673 * 10**9 # Hz
c = 3 * 10**8 # m/s
h = 6.626 * 10**-34 # J.s
k = 1.38 * 10**-23 # J/Kelvin
solar_mass = 1.989 * 10**30 #solar mass in kg

# Core Size

import numpy as np
import pandas as pd

# def Reff(theta_major, theta_minor):
#     return 0.01063 * np.sqrt(np.sqrt((theta_major**2 - theta_b**2) * (theta_minor**2 - theta_b**2))) * 206265 #/ (4 * np.log10(2))) # area in arcsec to parsec using conversion rule

def Reff(sigma_maj, sigma_min):
    return ((neta) * ((sigma_maj)*(sigma_min))**0.5)
    
neta = 1
theta_b = 0
theta_major = (data['major sigma (arcsec)'])
theta_minor = (data['minor sigma (arcsec)'])

sigma_maj = theta_major/(8 * np.log(2))**0.5
sigma_min = theta_minor/(8 * np.log(2))**0.5

# d_reff = 1 #  2.2 * 10**3# pc

# Surface and Volume Density

def surface_density(Mcore, Reff):
    return ((Mcore(d, Fv, Kv, freq, Td)) * solar_mass * ((10**3)/(3.086 * 10**18)**2)) / (np.pi * (Reff(sigma_maj, sigma_min))**2)

def mass_density(Mcore, Reff):
    return ((3 * Mcore(d, Fv, Kv, freq, Td)) * solar_mass * ((10**3)/ (3.086 * 10**18)**3)) / (4 * np.pi * (Reff(sigma_maj, sigma_min))**3)

# def number_density(Reff):
    # return (3 * 10**57)/ (((3.086 * 10**18)**3) * (4 * np.pi * (Reff(d, theta_major, theta_minor))**3))

def number_density(Mcore, Reff):
    return (((3 * Mcore(d, Fv, Kv, freq, Td)) * solar_mass * ((10**3)/ (3.086 * 10**18)**3)) / (4 * np.pi * (Reff(sigma_maj, sigma_min))**3))/(2.33 * 1.6735575 * 10**-24)

# Updating OLD table to NEW  OUTPUT table


data[r'Mass (Mo)'] = Mcore(d, Fv, Kv, freq, Td)
data['Radius (pc)'] = Reff(sigma_maj, sigma_min)
data['Surface Density (g/cm^2)'] = surface_density(Mcore, Reff)
data['Number Density (/cm^3)'] = number_density(Mcore, Reff)


# data.to_csv(r"C:\Users\Ayush Shivkumar\OneDrive\Desktop\PRL\Astrodendro Data\NEWastrodendro_catalogue.csv", index=False)
data.to_csv(r"C:\Users\Ayush Shivkumar\OneDrive\Desktop\PRL\Astrodendro Data\final_table_1.7kpc.csv", index=False)

print(BB(freq, Td))
print("\n")
print("The Core fluxes from Astrodendro are: \n{} ".format(Fv))
print("\n")
print("The core masses are: \n{} " .format(Mcore(d, Fv, Kv, freq, Td)))
print("\n")
print("The core sizes are: \n{} " .format(Reff(sigma_maj, sigma_min)))
print("\n")
print("The surface densities are: \n{} " .format(surface_density(Mcore, Reff)))
print("\n")
#print("The mass desnities are: \n{} " .format(mass_density(Mcore, Reff)))
#print("\n")
print("The number densities are: \n{} " .format(number_density(Mcore, Reff)))
```
